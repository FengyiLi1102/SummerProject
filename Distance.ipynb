```python
train_df_full = pd.read_csv(r'\\icnas4.cc.ic.ac.uk\fl4718\Desktop\Machine learning\Data\train.csv', index_col=0)
structures_df_full = pd.read_csv(r'\\icnas4.cc.ic.ac.uk\fl4718\Desktop\Machine learning\Data\structures.csv')
```

```python
def merge(df_1, df_2, indx):
    df_merge = pd.merge(df_1, df_2, how='left',
                       left_on=['molecule_name', f'atom_index_{indx}'],
                       right_on=['molecule_name', 'atom_index'])
    df_merge = df_merge.drop(['atom_index'], axis=1)
    df_merge = df_merge.rename(columns={'atom': f'atom_{indx}',
                                       'x': f'x_{indx}',
                                       'y': f'y_{indx}',
                                       'z': f'z_{indx}'})
    
    return df_merge
```

```python
train = merge(train_df_full, structures_df_full, 0)
train = merge(train, structures_df_full, 1)

test = merge(test_df_full, structures_df_full, 0)
test = merge(test, structures_df_full, 1)

train.head()
```

```python
%%time
# Calculate the distance
train_posi_0 = train[['x_0', 'y_0', 'z_0']].values
train_posi_1 = train[['x_1', 'y_1', 'z_1']].values
test_posi_0 = test[['x_0', 'y_0', 'z_0']].values
test_posi_1 = test[['x_1', 'y_1', 'z_1']].values

train['distance'] = np.linalg.norm(train_posi_0 - train_posi_1, axis=1)
test['distance'] = np.linalg.norm(test_posi_0 - test_posi_1, axis=1)
```
